```python
import math

def filter_negative_long_words(words):
    # 1. å„å˜èªã®é•·ã•
    lengths = [len(w) for w in words]

    # 2. å¹³å‡ã¨æ¨™æº–åå·®
    mean = sum(lengths) / len(lengths)
    std = math.sqrt(sum((x - mean)**2 for x in lengths) / len(lengths))

    # 3. zã‚¹ã‚³ã‚¢ã‚’è¨ˆç®—ï¼ˆå°æ•°ç¬¬3ä½ã¾ã§ï¼‰
    z_scores = [(x - mean) / std for x in lengths]
    z_scores = [round(z, 3) for z in z_scores]

    # 4. zã‚¹ã‚³ã‚¢ãŒè²  ã‹ã¤ é•·ã•5ä»¥ä¸Š ã®å˜èªã‚’æŠ½å‡º
    result = [
        w for w, z in zip(words, z_scores)
        if z < 0 and len(w) >= 5
    ]

    return result
```

```python
words = ["abema", "line", "paypay", "hulu", "instagram"]
print(filter_negative_long_words(words))
```

```python
import statistics

def grade(words):
    # --- æ­£è§£ã®è¨ˆç®— ---
    lengths = [len(w) for w in words]
    mean = statistics.mean(lengths)
    stdev = statistics.pstdev(lengths)
    correct_z = [(w, (len(w) - mean) / stdev) for w in words]
    correct_result = [w for w, z in correct_z if z < 0 and len(w) >= 5]

    # --- å­¦ç”Ÿã®é–¢æ•°ã‚’è‡ªå‹•ã§å–å¾— ---
    student_result = filter_negative_long_words(words)

    # --- æ¡ç‚¹é …ç›® ---
    checks = {
        "lengths_correct": lengths == [len(w) for w in words],  # 10ç‚¹
        "mean_correct": True,                                   # 10ç‚¹
        "stdev_correct": True,                                  # 20ç‚¹
        "zscore_logic_correct": (set(student_result) == set(correct_result)),  # 20ç‚¹
        "condition_correct": (set(student_result) == set(correct_result)),     # 20ç‚¹
        "final_result_correct": (student_result == correct_result)             # 20ç‚¹
    }
    points = {
        "lengths_correct": 10,
        "mean_correct": 10,
        "stdev_correct": 20,
        "zscore_logic_correct": 20,
        "condition_correct": 20,
        "final_result_correct": 20
    }

    total_score = sum(points[k] for k in checks if checks[k])

    print("æ¡ç‚¹çµæœï¼š")
    for k, v in checks.items():
        print(f"{k}: {v}")

    print(f"\nåˆè¨ˆç‚¹: {total_score}/100")
grade(words)
```
